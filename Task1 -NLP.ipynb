```python
# Task 2:Build Advanced Preprocessing Function

import re

def preprocess_text(text):
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1+', r'\1\1', text)
    text = re.sub(r'\s+', ' ', text).strip()
    text = text.lower()
    tokens = text.split()
    tokens = [word for word in tokens if len(word) > 2 or word in ['i', 'no', 'not']]
    text = ' '.join(tokens)
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'\S+@\S+', '', text)
    return text
```

```python
# Example
text ="I have 2 dogs"
result = preprocess_text(text)
print(result)
```

```python
# Task 3: Stress Testing
import re

def clean_text(sentences):
    cleaned_data = []
    for sentence in sentences:
        original_text = sentence
        cleaned_tokens = re.findall(r'\b\w+\b', sentence)
        cleaned_sentence = ' '.join(cleaned_tokens)
        cleaned_data.append((original_text, cleaned_tokens, cleaned_sentence))
    return cleaned_data

n = int(input("Enter number of sentences: "))

sample_inputs = []
for i in range(n):
    text = input(f"Enter sentence {i+1}: ")
    sample_inputs.append(text)

output = clean_text(sample_inputs)

for original_text, cleaned_tokens, cleaned_sentence in output:
    print(f"Original Text: {original_text}")
    print(f"Cleaned Tokens: {cleaned_tokens}")
    print(f"Cleaned Sentence: {cleaned_sentence}\n")
```

```python
# Task 4: Token Analytics
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('punkt_tab')

results = []

n = int(input("Enter number of sentences: "))

sentences = []
for i in range(n):
    sentence = input(f"Enter sentence {i+1}: ")
    sentences.append(sentence)

for sentence in sentences:
    tokens = word_tokenize(sentence)
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    if total_tokens > 0:
        avg_token_length = sum(len(token) for token in tokens) / total_tokens
    else:
        avg_token_length = 0

    results.append((sentence, total_tokens, unique_tokens, avg_token_length))

most_noise_sentence = max(results, key=lambda x: x[1])[0]
most_meaningful_sentence = min(results, key=lambda x: x[2])[0]

for sentence, total, unique, avg_len in results:
    print("\nSentence:", sentence)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", round(avg_len, 2))

print("\nSentence with Most Noise (Max Tokens):", most_noise_sentence)
print("Most Meaningful Sentence (Min Unique Tokens):", most_meaningful_sentence)
```

```python
# Task 5: Frequency Analysis
from collections import Counter

n = int(input("Enter number of words: "))

tokens = []
for i in range(n):
    word = input(f"Enter word {i+1}: ")
    tokens.append(word)

word_counts = Counter(tokens)

top_10_frequent = word_counts.most_common(10)
top_5_least_frequent = word_counts.most_common()[:-6:-1]

print("Top 10 Most Frequent Words:", top_10_frequent)
print("Top 5 Least Frequent Words:", top_5_least_frequent)
```

```python
# Task 6: Build Full Pipeline
def full_pipeline(text_list):
    tokens = []
    clean_sentences = []
    for sentence in text_list:
        tokenized = sentence.split()
        tokens.extend(tokenized)
        clean_sentences.append(' '.join(tokenized))
    return {
        "tokens": tokens,
        "clean_sentences": clean_sentences
    }

n = int(input("Enter number of sentences: "))

text_list = []
for i in range(n):
    sentence = input(f"Enter sentence {i+1}: ")
    text_list.append(sentence)

result = full_pipeline(text_list)

print("Tokens:", result["tokens"])
print("Clean Sentences:", result["clean_sentences"])
```

```python
# Task 7: Error Handling
import emoji

def analyze_string(input_string):
    if not input_string:
        return "Empty string"
    elif all(char in emoji.EMOJI_DATA for char in input_string):
        return "Only emojis"
    elif input_string.isdigit():
        return "Only numbers"
    else:
        return "Mixed content"

input_string = input("Enter a string: ")

result = analyze_string(input_string)
print(result)
```
