```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

# 1. Load Transformer Model
model_name = "microsoft/DialoGPT-medium"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

chat_history_ids = None

print("Chatbot: Hello! I am your AI assistant. How can I help you today?")

while True:
    # 2. User Input
    user_input = input(" user: ")

    # 5. Exit Condition
    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Chatbot ends the conversation")
        break

    user_lower = user_input.lower()

    # 3.  Responses
    if "hello" in user_lower:
        response = "Hello! Nice to meet you. How can I assist you today?"

    elif "artificial intelligence" in user_lower:
        response = "Artificial Intelligence refers to the simulation of human intelligence by machines that can perform tasks such as learning, reasoning, and problem solving."

    elif "who created python" in user_lower or "python creator" in user_lower:
        response = "Python was created by Guido van Rossum and released in 1991."

    elif "thank you" in user_lower or "thanks" in user_lower:
        response = "You're welcome! Feel free to ask more questions."

    else:
        # 4. Transformer-based Response (fallback)
        new_input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors='pt')

        if chat_history_ids is not None:
            input_ids = torch.cat([chat_history_ids, new_input_ids], dim=-1)
        else:
            input_ids = new_input_ids

        chat_history_ids = model.generate(
            input_ids,
            max_length=1000,
            pad_token_id=tokenizer.eos_token_id,
            do_sample=True,
            top_k=50,
            top_p=0.95,
            temperature=0.75
        )

        response = tokenizer.decode(
            chat_history_ids[:, input_ids.shape[-1]:][0],
            skip_special_tokens=True
        )

    # Output
    print(f"Chatbot: {response}")
```
